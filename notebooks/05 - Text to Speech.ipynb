# Generate API Keys

1. Follow Instructions here: https://docs.play.ht/reference/api-authentication
2. Store the credentials in your `.env` file
    ```
    PYHT_USER_ID = 'xxxx'
    PYHT_SECRET = 'xxxx'
    ```
3. Choose a voice from the list of voices and get the `id` of chosen voice. 
4. store `id` in the `.env` file
    ```
    PYHT_VOICE_NARRATOR = 's3://voice-cloning-zero-shot/b3def996-302e-486f-a234-172fa0279f0e/anthonysaad/manifest.json'
    ```
